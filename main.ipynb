```python
import os
from dotenv import load_dotenv
from google import genai
import sys
```

```python
from google.genai import types
```

Below loads the API key and initiates the client object

```python

load_dotenv()
api_key = os.environ.get("GEMINI_API_KEY")

client = genai.Client(api_key=api_key)
```

Collect the argv as the input to the agent

```python

if len(sys.argv) < 2:
    print("need a prompt")
    sys.exit(1)
print("Args", sys.argv)
#can give argv by doing: 
#uv run main.py "this is going to be argv"

#prompt = sys.argv[1]
prompt = "what is 10+9"
```

```python
messages = [
    types.Content(role="user", parts=[types.Part(text=prompt)]),
]
```

Calls the agent with the prompt using 'models.generate_content'

```python
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="what is 9 + 10",
    #contents = messages
)
```

prints out response and the properties of the response object

```python
print(response.text)

if response is None or response.usage_metadata is None:
    print("bad response")
else: 
    print(f"prompt tokens:{response.usage_metadata.prompt_token_count}")
    print(f"response tokens:{response.usage_metadata.candidates_token_count}")
```
